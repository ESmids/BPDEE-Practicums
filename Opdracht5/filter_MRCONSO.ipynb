```python
term2cuis = {}
with open("MRCONSO.RRF", "r", encoding="utf-8") as file:
    for regel in file:
        regel  = regel.strip()
        fields = regel.split("|")
        lat    = fields[1]
        ispref = fields[6]
        suprss = fields[16]
        cui    = fields[0]
        term   = fields[14]

        if lat == "ENG" and suprss == "N" and ispref == "Y" and term not in stopwords:
            term = term.lower()
            if len(term)<100 and len(term)>1:                
                if term not in term2cuis:
                    term2cuis[term] = set()
                cuis = term2cuis[term]
                cuis.add(cui)
```

```python
umlsterms = list(term2cuis.keys())
```

```python
import ahocorasick
import pickle

# 1️⃣ Woordenlijst (kan miljoenen woorden zijn)
#words = ["he", "she", "his", "hers"]
words = umlsterms

#automaton = ahocorasick.Automaton()
# 2️⃣ Maak de automaat
A = ahocorasick.Automaton()

# Voeg alle woorden toe
for idx, word in enumerate(words):
    A.add_word(word, (idx, word))  # tweede argument = info die je terug wilt

# Finaliseer de automaat (bouw failure links)
A.make_automaton()

# opslaan
with open("automation.pkl", "wb") as f:
    pickle.dump(A, f)
```
