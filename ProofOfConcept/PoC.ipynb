# PoC – Curriculum Assistant (LLM)

Dit is een simpele proof of concept voor:
- classificatie van leerdoelen (Bloom)
- detectie van misalignment
- genereren van feedback

Dit is een gesimplificeerde versie zonder echte LLM.

# 1. Data

```python
# Voorbeelddata (normaal zou dit uit documenten komen)
leerdoelen = [
    "De student kan de basisprincipes van machine learning uitleggen",
    "De student kan een regressiemodel implementeren in Python",
    "De student kan de resultaten analyseren en interpreteren"
]

toetsvormen = [
    "meerkeuzetoets",
    "programmeeropdracht",
    "verslag"
]
```

# 2. classificatie (Bloom)

```python
def classify_bloom(leerdoel):
    leerdoel = leerdoel.lower()
    
    if "uitleggen" in leerdoel:
        return "Begrijpen"
    elif "implementeren" in leerdoel:
        return "Toepassen"
    elif "analyseren" in leerdoel:
        return "Analyseren"
    else:
        return "Onbekend"
```

## Test

```python
for ld in leerdoelen:
    print(ld, "->", classify_bloom(ld))
```

# 3. Alignment check

```python
def check_alignment(bloom_level, toetsvorm):
    if bloom_level == "Begrijpen" and toetsvorm == "meerkeuzetoets":
        return "OK"
    elif bloom_level == "Toepassen" and toetsvorm == "programmeeropdracht":
        return "OK"
    elif bloom_level == "Analyseren" and toetsvorm == "verslag":
        return "OK"
    else:
        return "Mogelijke misalignment"
```

## Test

```python
for ld, tv in zip(leerdoelen, toetsvormen):
    level = classify_bloom(ld)
    result = check_alignment(level, tv)
    print(f"{ld} | {tv} -> {result}")
```

# 4. Feedback genereren

```python
def generate_feedback(leerdoel, toetsvorm, alignment):
    if alignment == "OK":
        return "Toetsvorm sluit goed aan op leerdoel."
    else:
        return f"Overweeg een andere toetsvorm voor: '{leerdoel}'"
```

## Test

```python
for ld, tv in zip(leerdoelen, toetsvormen):
    level = classify_bloom(ld)
    alignment = check_alignment(level, tv)
    feedback = generate_feedback(ld, tv, alignment)
    
    print("Leerdoel:", ld)
    print("Toets:", tv)
    print("Bloom:", level)
    print("Resultaat:", alignment)
    print("Feedback:", feedback)
    print("-" * 40)
```

# 5. tabel output

```python
import pandas as pd

data = []

for ld, tv in zip(leerdoelen, toetsvormen):
    level = classify_bloom(ld)
    alignment = check_alignment(level, tv)
    feedback = generate_feedback(ld, tv, alignment)
    
    data.append({
        "Leerdoel": ld,
        "Bloom": level,
        "Toetsvorm": tv,
        "Alignment": alignment,
        "Feedback": feedback
    })

df = pd.DataFrame(data)
df
```
